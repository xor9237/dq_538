We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.
- etc.

***1. Do students in more popular majors make more money?***

```python
recent_grads.sort_values(by='Total',ascending=False).plot(x='Total', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

--> Seemingly popular major does not guarantee higher wage since it shows majors with higher number of students earn low income while many majors with lesser students make more money.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

***2. Do students that majored in subjects that were majority female make more money?***

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

---> Females that majored in the popular majors are not making more money

***3. Is there any link between the number of full-time employees and median salary?***

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

--> There is higher chance of making more salary by majoring in less popular majors.

***Explore the distribution of values in each column.***

```python
import matplotlib.pyplot as plt

columns = ['Sample_size', 'Median', 'Employed', 'Full_time',
          'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(5,35))
num=1
for col in columns:
    ax = fig.add_subplot(8,1,num)
    ax = recent_grads[col].plot(kind='hist')
    ax.set_xlabel(col)
    num+=1
```
